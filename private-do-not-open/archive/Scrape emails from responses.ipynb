```python
subfolders = [ f.path for f in os.scandir(root) if f.is_dir() ]
# subfolders
```

```python
import re 

emails_all = set()
files = []
for subfolder in subfolders:
    if any(subfolder.endswith(suffix) for suffix in ['ipynb_checkpoints','.git','Archive']):
        continue
#     print(subfolder)
    subfiles = [f.path for f in os.scandir(subfolder)]
    files += subfiles

for file in files:
#     print(file)
    try:
        with open(file) as f: # Use file to refer to the file object
            text = f.read()
        emails_curr = re.findall(r'[\w.+-]+@[\w-]+\.[\w.-]+', text)
        print(len(emails_curr))
        emails_all = emails_all.union(set(emails_curr))
    except Exception as e:
        print(e)
```
