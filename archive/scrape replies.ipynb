```python
import os
root = os.getcwd()
root = os.path.join(root, 'replies')
print(root)
```

```python
subfolders = [ f.path for f in os.scandir(root) if f.is_dir() ]
print(subfolders)
```

```python
# import re 

# emails_all = set()
files = []
for subfolder in subfolders:
    if any(subfolder.endswith(suffix) for suffix in ['ipynb_checkpoints','.git','Archive']):
        continue
    # print(subfolder)
    subfiles = [f.path for f in os.scandir(subfolder)]
    files += subfiles
```

```python
my_file = files[0]
with open(my_file) as f: # Use file to refer to the file object
    text = f.read()
    text_anon = ''
    record = True
    for char in text:
        if char == '<':
            record = False
        if record:
            text_anon += char
        if char == '>':
            record = True
print(text_anon)
```

```python
replies = []

for file in files:
    # print(file)
    try:
        with open(file) as f: # Use file to refer to the file object
            text = f.read()
            # text = text[71:]
            # text = text[71:] + text[456:]
            reply = ''
            record = True
            for char in text:
                if char == '<':
                    record = False
                if record:
                    reply += char
                if char == '>':
                    record = True    
            replies.append(reply)
            # print(len(text))
    except Exception as e:
        # print(e)
        continue
```
